# Desafio 6:
Analise mais detalhadamente o gráfico pairplot, gaste um tempo pensando e tentando enteder os gráficos.

```python
sns.pairplot(data = imdb_usa[['gross', 'budget', 'lucro', 'title_year']])
plt.show()
```

```python
imdb_usa[['gross', 'budget', 'lucro', 'title_year']].corr()
```

O gráfico mostra correlações positivas entre gross-budget e gross-lucro. As outras correlações estão classificadas como bem-fracas ou fracas (conforme a definição da tabela de correlação linear). 

Devemos considerar que a relação entre gross - budget e gross - lucro pode ser devido ao fato que diretores ou estudios mais famosos possuem, de modo geral, um orçamento maior e mais fãs, possibilitando que disponham de mais dinheiro para suas produções e ainda assim consigam recuperar o investido e lucrar em cima.

# desafio 7
calcular a correlação dos filmes para depois de 2000

```python
filmes_pos_2000 = imdb_usa[['gross', 'budget', 'lucro', 'title_year']].loc[imdb_usa['title_year'] > 2000]
filmes_pos_2000.corr()
```

Em comparação à correlação anterior vista, para os filmes de depois de 2000 a correlação entre gross-budget e budget-lucro aumentou enquanto entre gross-lucro diminuiu, todos com pequenas margens de diferenças.

Olhando para budget-lucro, foi um aumento pequeno no valor da correlação desses filmes do imdb, mas que condiz com o esperado de que o budget é cada vez maior com o tempo assim como o lucro, os dois crescendo de forma bem semelhante.

# desafio 8

tentar encontrar uma reta no gráfico que parece se aproximar com uma reta (por exemplo budget/lucro, gross/lucro)

```python
sns.lmplot(x ='gross', y='budget', data = imdb_usa)
plt.title('Gross x Budget')
plt.ylabel('Gross')
plt.xlabel('Budget')

sns.lmplot(x ='lucro', y='budget', data = imdb_usa)
plt.title('Budget x Lucro')
plt.ylabel('Budget')
plt.xlabel('Lucro')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(imdb_usa, y_vars='lucro', x_vars=['budget', 'gross'], height=6, kind='reg')

plt.show()
```
